### Plot functions in pandas

```python
import matplotlib as plt
```

```python
data['season'].value_counts().plot(kind='bar') # Vertical bar plot. Great for comparing discrete categories.
```

```python
data['season'].value_counts().plot(kind='line') # Line plot (the default type if no kind is specified). Ideal for showing trends over time.
```

```python
data['season'].value_counts().plot(kind='barh') #Horizontal bar plot. Useful for long categorical names.
```

```python
data['season'].value_counts().plot(kind='hist') #Histogram. Used to look at frequency distributions of numerical values.
```

```python
data['season'].value_counts().plot(kind='box') #Boxplot. Summarizes dataset distribution using five-number summaries (quartiles and outliers).
```

```python
data['season'].value_counts().plot(kind='kde')   #Kernel Density Estimation plot. Represents probability density as a smooth curve.
```

```python
data['season'].value_counts().plot(kind='density') #Kernel Density Estimation plot. Represents probability density as a smooth curve.
```

```python
data['season'].value_counts().plot(kind='area') #Area plot. A filled version of a line graph displaying cumulative totals.
```

```python
data['season'].value_counts().plot(kind='pie')
```
